```python
from qiskit.circuit.library import TwoLocal, EfficientSU2,RealAmplitudes
import matplotlib.pyplot as plt
```

```python
# Create a TwoLocal ansatz circuit
two_local_ansatz = TwoLocal(num_qubits=4, reps=2, rotation_blocks='ry', entanglement_blocks='cz', entanglement='linear')
two_local_ansatz.decompose().draw(output='text', fold=-1)
```

```python
two_local_ansatz = TwoLocal(num_qubits=4, reps=2, rotation_blocks='ry', entanglement_blocks='cx', entanglement='circular')
two_local_ansatz.decompose().draw(output='text', fold=-1)
```

```python
efficient_su2_ansatz = EfficientSU2(num_qubits=3, reps=2, entanglement='linear')
efficient_su2_ansatz.decompose().draw(output='text', fold=-1)
```

```python
real_ampl_ansatz=RealAmplitudes(num_qubits=4, reps=2, entanglement='linear')
real_ampl_ansatz.decompose().draw(output='text', fold=-1)
```
